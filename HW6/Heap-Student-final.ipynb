# Heap Class

```python
class Heap:
    def __init__(self, isMinHeap=True):
        self._heap = [None] 
        self._pos = 0
        self._minheap = isMinHeap
    
    def size(self) -> int:
        print("Remove this line and write code")
    
    
    def is_empty(self) -> bool:
        return self._pos == 0
    
    def left(self, x: int) -> int:
        n = x * 2
        return n if n <= self._pos else None
    
    def right(self, x: int) -> int:
        n = x * 2 + 1
        return n if n <= self._pos else None
    
    def father(self, x: int) -> int:
        return x // 2 if x > 1 else None

    def getTop(self) -> int:
        if not self._heap:
            raise IndexError("Heap is empty")
        return self._heap[1]
    
        
    
    def buildHeap(self, arr: list):
        self._heap = [None] + arr
        self._pos = len(arr)
        for i in range(self._pos // 2, 0, -1):
            self._heapify_down(i)

    def insert(self, x: int):
        self._heap.append(x)
        self._pos += 1
        self._heapify_up(self._pos)

    def deleteTop(self):
        if self.is_empty():
            return None
        top = self.getTop()
        self._heap[1], self._heap[self._pos] = self._heap[self._pos], self._heap[1]
        self._heap.pop()
        self._pos -= 1
        if self._minheap:
            self._heapify_down_min(1)
        else:
            self._heapify_down_max(1)
        return top
            
    def _heapify_up(self, index):
        while index > 1:
            parent = self.father(index)
            if self._minheap:
                if self._heap[parent] > self._heap[index]:
                    self._heap[parent], self._heap[index] = self._heap[index], self._heap[parent]
                    index = parent
                else:
                    break
            else:
                if self._heap[parent] < self._heap[index]:
                    self._heap[parent], self._heap[index] = self._heap[index], self._heap[parent]
                    index = parent
                else:
                    break

    def _heapify_down(self, index):
        while index * 2 <= self._pos:
            child = self.left(index)
            if child < self._pos:
                if self._minheap:
                    if self._heap[child] > self._heap[child + 1]:
                        child += 1
                else:
                    if self._heap[child] < self._heap[child + 1]:
                        child += 1

            if self._minheap:
                if self._heap[index] > self._heap[child]:
                    self._heap[index], self._heap[child] = self._heap[child], self._heap[index]
                    index = child
                else:
                    break
            else:
                if self._heap[index] < self._heap[child]:
                    self._heap[index], self._heap[child] = self._heap[child], self._heap[index]
                    index = child
                else:
                    break
                
    def _heapify_down_min(self, index):
        while True:
            left = self.left(index)
            right = self.right(index)
            smallest = index

            if left and self._heap[left] < self._heap[smallest]:
                smallest = left
            if right and self._heap[right] < self._heap[smallest]:
                smallest = right

            if smallest == index:
                break

            self._heap[index], self._heap[smallest] = self._heap[smallest], self._heap[index]
            index = smallest

    def _heapify_down_max(self, index):
        while True:
            left = self.left(index)
            right = self.right(index)
            largest = index

            if left and self._heap[left] > self._heap[largest]:
                largest = left
            if right and self._heap[right] > self._heap[largest]:
                largest = right

            if largest == index:
                break

            self._heap[index], self._heap[largest] = self._heap[largest], self._heap[index]
            index = largest

    def write_dot(self, filename):
        with open(filename, 'w') as f:

            f.write("digraph Heap {\n")
            f.write("    node [shape=record];\n")

            for i in range(1, self._pos + 1):
                f.write(f'    {i} [label="{i}|{self._heap[i]}"];\n')

            for i in range(1, self._pos + 1):
                left_child = self.left(i)
                right_child = self.right(i)

                if left_child is not None and left_child <= self._pos:
                    f.write(f'    {i} -> {left_child};\n')  # No color, black edge by default
                if right_child is not None and right_child <= self._pos:
                    f.write(f'    {i} -> {right_child};\n')  # No color, black edge by default

            f.write('}\n')
        
    
```
